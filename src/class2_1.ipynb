# 离散傅里叶变换的实现与显示

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("../img/lena_gray.jpg",0) #以灰度图方式读入
img1.shape
```

定义离散傅里叶变换函数

```python
def DFT(img):
    h,w = img.shape
    dft = np.zeros((h,w)).astype(complex)
    x = np.arange(w).repeat(h).reshape(w,-1).transpose()
    y = np.arange(h).repeat(w).reshape(h,-1)
    for v in range(h):
        print(v,"iter",end = " ")
        for u in range(w):
            e = np.exp(-2j*np.pi*(x*u/w+y*v/h))
            dft[v,u] = np.sum(img*np.power(-1,x+y)*e)
    return np.abs(dft)
```

将512*512的图像缩放为100*100，否则运算速度太慢

```python
img100 = cv2.resize(img1,(100,100))
img100.shape
```

```python
dft = DFT(img100)
```

```python
fig = plt.figure()
ax = fig.add_subplot(121)
plt.imshow(dft,cmap="gray")
plt.colorbar(shrink = 0.8)
ax = fig.add_subplot(122)
plt.imshow(np.log(dft),cmap="gray")
plt.colorbar(shrink = 0.8)
```
